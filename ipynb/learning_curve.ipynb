```python
from lightjob.cli import load_db
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
db  = load_db()
```

```python
jobs = db.jobs_with(where='jobset41', state='success')
```

```python
for j in jobs:
    df = pd.read_csv('../jobs/results/{}/csv/stats.csv'.format(j['summary']))
    #plt.plot(df['duration'].cumsum()/3600., df['avg_loss_train_fix'])
    plt.plot(df['train_recons_error'].dropna())
    print(df['train_recons_error'].dropna().argmin())
    plt.show()
```
